```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
relu = np.load("data/relu.npy")
sigmoid = np.load("data/sigmoid.npy")
tanh = np.load("data/tanh.npy")
```

```python
plt.title('Losses Comparision')

plt.plot(relu[0], label='ReLU', color='blue')
plt.plot(sigmoid[0], label='Sigmoid', color='red')
plt.plot(tanh[0], label='Tanh', color='green')

plt.legend()
plt.savefig('plots/losses_comparision.png', dpi=300)
plt.show()
```

```python
# Shape: (Num Trials, Num Epochs)

# Find best and worst loss per activation at each epoch
relu_mean = np.array([np.mean(relu[:, i]) for i in range(relu.shape[1])])
relu_best = np.array([np.min(relu[:, i]) for i in range(relu.shape[1])])
relu_worst = np.array([np.max(relu[:, i]) for i in range(relu.shape[1])])

sigmoid_best = np.array([np.min(sigmoid[:, i]) for i in range(sigmoid.shape[1])])
sigmoid_worst = np.array([np.max(sigmoid[:, i]) for i in range(sigmoid.shape[1])])
sigmoid_mean = np.array([np.mean(sigmoid[:, i]) for i in range(sigmoid.shape[1])])

tanh_best = np.array([np.min(tanh[:, i]) for i in range(tanh.shape[1])])
tanh_worst = np.array([np.max(tanh[:, i]) for i in range(tanh.shape[1])])
tanh_mean = np.array([np.mean(tanh[:, i]) for i in range(tanh.shape[1])])
```

```python
plt.title('Mean Loss Comparision')

plt.fill_between(range(len(relu_mean)), relu_best, relu_worst, alpha=0.1, color='blue')
plt.fill_between(range(len(sigmoid_mean)), sigmoid_best, sigmoid_worst, alpha=0.1, color='red')
plt.fill_between(range(len(tanh_mean)), tanh_best, tanh_worst, alpha=0.1, color='green')

plt.plot(relu_mean, label='ReLU', color='blue')
plt.plot(sigmoid_mean, label='Sigmoid', color='red')
plt.plot(tanh_mean, label='Tanh', color='green')


plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.savefig('plots/mean_losses_comparision.png', dpi=300)
plt.show()
```
